```python
n_neighbors = list(range(1, 8))
```

```python
results_plot = list()

X_train, X_test, y_train, y_test = train_test_split(np_feature_array, label_list)
for i in n_neighbors:
    neigh = KNeighborsRegressor(n_neighbors=i)
    neigh.fit(X_train, y_train)

    pred_value = list()
    pred = neigh.predict(X_test)
    m = mean_squared_error(pred, y_test)
    results_plot.append(m)
#     print(pred)
```
